```python
# 辅助函数
import numpy as np
def loaddata(filename):
    data = []
    label = []
    with open(filename) as fr:
        for line in fr.readlines():
            arr = line.strip().split('\t')
            data.append([float(arr[0]),float(arr[1])])
            label.append(float(arr[2]))
    datama = np.array(data)
    labelma = np.expand_dims(np.array(label),axis=1)# (n,1)维数组
    return datama,labelma
```

```python
def selj(i,m):
    j=i
    while(j==i):
        j = int(np.random.uniform(0,m))
    return j
def clipa(aj,H,L):
    if aj>H: aj=H
    if L>aj: aj=L
    return aj
def lh(x,y,labeli,labelj,C):
    if (labeli != labelj):
        L = max(0,y-x)
        H = min(C,C+y-x)
    else:
        L = max(0,y+x-C)
        H = min(C,y+x)
    return L,H
        
    
```

```python
def smosim(data,label,C,toler,num):
    datama = np.array(data)
    labelma = np.expand_dims(np.array(label),axis=1)# (n,1)维数组
    b = 0
    m,n = datama.shape
    alphas = np.zeros((m,1))
    cnt = 0
    while (cnt<num):#迭代次数，当有一次修改就置为0，直到num次alchanged都为0
        alchanged = 0
        for i in range(m):
            wt = np.dot(np.multiply(alphas,labelma).T,datama)#multiply((100,1),(100,1)).T =dot((1,100),100,2) = (1,2)
            fxi = np.dot(wt,datama[i,:].T)+b
            ei = fxi - labelma[i]
            if ((labelma[i]*ei<-toler)and(alphas[i]<C)) or ((labelma[i]*ei>toler)and(alphas[i]>0)):
                j = selj(i,m)
                fxj = np.dot(wt,datama[j,:].T)+b
                ej = fxj - labelma[j]
                iold = alphas[i].copy()
                jold = alphas[j].copy()
                L,H = lh(alphas[i],alphas[j],labelma[i],labelma[j],C)
                if L==H: 
                    print('L=H')
                    continue
                eta = 2.0*np.dot(datama[i,:],datama[j,:].T)-np.dot(datama[i,:],datama[i,:].T)-np.dot(datama[j,:],datama[j,:].T)
                if eta>=0: 
                    print('eta>=0')
                    continue
                alphas[j] -= labelma[j]*(ei-ej)/eta
                alphas[j] = clipa(alphas[j],H,L)
                if abs(jold-alphas[j])<0.00001:
                    print('not moving enough')
                    continue
                alphas[i] += labelma[i]*labelma[j]*(jold-alphas[j])
                b1 = b - ei - labelma[i]*(alphas[i]-iold)*np.dot(datama[i,:],datama[i,:].T)-labelma[j]*(alphas[j]-jold)*np.dot(datama[i,:],datama[j,:].T)
                b2 = b - ej - labelma[i]*(alphas[i]-iold)*np.dot(datama[i,:],datama[j,:].T)-labelma[j]*(alphas[j]-jold)*np.dot(datama[j,:],datama[j,:].T)
                if (0<alphas[i]) and (C>alphas[i]): b = b1
                elif (0<alphas[j]) and (C>alphas[j]): b = b2
                else: b = (b1+b1)/2.0
                alchanged += 1
                print('iter:%d i:%d pairs changed %d' %(cnt,i,alchanged))
        if (alchanged==0): cnt +=1
        else: cnt = 0
        print('iteration number: %d' %cnt)
    return b,alphas
            
```

```python
class Optstruct(object):
    def __init__(self,data,label,C,toler):
        self.x = data
        self.label = label
        self.C = C
        self.tol = toler
        self.m = data.shape[0]
        self.alphas = np.zeros((self.m,1))
        self.b = 0
        self.ecache = np.zeros((self.m,2))#误差缓存
```

```python
def calcek(opt,i):
    wt = np.dot(np.multiply(opt.alphas,opt.label).T,opt.x)#multiply((100,1),(100,1)).T =dot((1,100),100,2) = (1,2)
    fxi = np.dot(wt,opt.x[i,:].T) + opt.b
    #print(fxi)
    ei = fxi - opt.label[i]
    return ei
```

```python
def selectj(opt,i,ei):
    maxk = -1
    maxde = 0
    ej = 0
    opt.ecache[i] = [i,ei]#将i塞到误差缓存中表示计算过了
    validlist = np.nonzero(opt.ecache[:,0])[0]#nonzero会生成一个对应dim的tuple，tuple内包含array，这里选取第一个array，也只有一个array
    if len(validlist)>1:
        for k in validlist:
            if k==i:
                continue
            ek = calcek(opt,k)
            de = abs(ek-ei)
            if de > maxde:
                maxde = de
                maxk = k
                ej = ek
        return maxk,ej
    else:
        j = selj(i,opt.m)
        ej = calcek(opt,j)
        return j,ej
def updateek(opt,k):
    ek = calcek(opt,k)
    opt.ecache[k,:] = [1,ek]    
    
    
    
```

```python
def inner(i,opt):
    ei = calcek(opt,i)
    #print(ei.shape)
    if ((opt.label[i,0]*ei < -opt.tol)and(opt.alphas[i,0] < opt.C)) or ((opt.label[i,0]*ei>opt.tol)and(opt.alphas[i,0]>0)):
        j,ej = selectj(opt,i,ei)
        iold = opt.alphas[i].copy()
        jold = opt.alphas[j].copy()
        L,H = lh(opt.alphas[i],opt.alphas[j],opt.label[i],opt.label[j],opt.C)
        if L==H:
            print('L=H')
            return 0
        eta = 2.0*np.dot(opt.x[i,:],opt.x[j,:].T)-np.dot(opt.x[i,:],opt.x[i,:].T)-np.dot(opt.x[j,:],opt.x[j,:].T)
        if eta >= 0: 
            print('eta>=0')
            return 0
        opt.alphas[j] -= opt.label[j]*(ei-ej)/eta
        opt.alphas[j] = clipa(opt.alphas[j],H,L)
        updateek(opt,j)
        if abs(opt.alphas[j] - jold) < 0.00001:
            print('j not moving enough')
            return 0
        opt.alphas[i] += opt.label[j]*opt.label[i]*(jold - opt.alphas[j])
        updateek(opt,i)
        b1 = opt.b - ei - opt.label[i]*(opt.alphas[i]-iold)*np.dot(opt.x[i,:],opt.x[i,:].T)-opt.label[j]*(opt.alphas[j]-jold)*np.dot(opt.x[i,:],opt.x[j,:].T)
        b2 = opt.b - ej - opt.label[j]*(opt.alphas[j]-iold)*np.dot(opt.x[i,:],opt.x[j,:].T)-opt.label[j]*(opt.alphas[j]-jold)*np.dot(opt.x[j,:],opt.x[j,:].T)
        if (0<opt.alphas[i]) and (opt.alphas[i]<opt.C): opt.b = b1
        elif (0<opt.alphas[j]) and (opt.alphas[j]<opt.C): opt.b = b2
        else: opt.b = (b1+b2)/2
        return 1
    else:
        return 0
            
```

```python
def smo(datama,labelma,C,toler,num,ktup=('lin',0)):
    opt = Optstruct(datama,labelma,C,toler)
    cnt = 0
    entire = True
    achanged = 0
    while(cnt<num) and ((achanged>0) or (entire)):#achanged=0 and entire为False 循环结束
        achanged = 0
        if entire:
            for i in range(opt.m):
                achanged += inner(i,opt)
                print('fullset,iter:%d i:%d,pairs changed %d' %(cnt,i,achanged))
            cnt +=1
        else:
            nonbou = np.nonzero((opt.alphas>0)*(opt.alphas<opt.C))[0]
            for i in nonbou:
                achanged += inner(i,opt)
                print('nonbound,iter:%d i:%d pairs changed %d' %(cnt,i,achanged))
            cnt +=1
        if entire: entire = False
        elif (achanged==0): entire=True
        print('iteration number:%d' %cnt)
    return opt.b,opt.alphas
            
```

**应用复杂核函数**
1.kernelT 核函数

```python
#在复杂数据上应用核函数
def kernelT(X,A,ktup):
    m,n = X.shape
    if ktup[0]=='lin': K = np.dot(X,A.T)
    elif ktup[0]=='rbf':
        XX = X-A
        K = np.dot(XX,XX.T).diagonal()#获取对角线元素生成一个一维数组
        #K = K.reshape(m,1)
        K = np.exp(K/(-1*ktup[1]**2))
    else: raise NameError('The Kernel is not recognized')
    return K        
```

```python
class Opt(object):
    def __init__(self,data,label,C,toler,ktup):
        self.x = data
        self.label = label
        self.C = C
        self.tol = toler
        self.m = data.shape[0]
        self.alphas = np.zeros((self.m,1))
        self.b = 0
        self.ecache = np.zeros((self.m,2))#误差缓存
        self.K = np.zeros((self.m,self.m))
        for i in range(self.m):
            self.K[:,i] = kernelT(self.x,self.x[i,:],ktup)#全局的K只计算一次        
```

```python
def selectj(opt,i,ei):
    maxk = -1
    maxde = 0
    ej = 0
    opt.ecache[i] = [i,ei]#将i塞到误差缓存中表示计算过了
    validlist = np.nonzero(opt.ecache[:,0])[0]#nonzero会生成一个对应dim的tuple，tuple内包含array，这里选取第一个array，也只有一个array
    if len(validlist)>1:
        for k in validlist:
            if k==i:
                continue
            ek = calcEk(opt,k)
            de = abs(ek-ei)
            if de > maxde:
                maxde = de
                maxk = k
                ej = ek
        return maxk,ej
    else:
        j = selj(i,opt.m)
        ej = calcEk(opt,j)
        return j,ej
def updateek(opt,k):
    ek = calcEk(opt,k)
    opt.ecache[k,:] = [1,ek] 
    
```

```python
#将向量内积换成核函数
def innerL(i,opt):
    ei = calcEk(opt,i)
    #print(ei.shape)
    if ((opt.label[i,0]*ei < -opt.tol)and(opt.alphas[i,0] < opt.C)) or ((opt.label[i,0]*ei>opt.tol)and(opt.alphas[i,0]>0)):
        j,ej = selectj(opt,i,ei)
        iold = opt.alphas[i].copy()
        jold = opt.alphas[j].copy()
        L,H = lh(opt.alphas[i],opt.alphas[j],opt.label[i],opt.label[j],opt.C)
        if L==H:
            print('L=H')
            return 0
        eta = 2.0*opt.K[i,j]-opt.K[i,i]-opt.K[j,j]
        if eta >= 0: 
            print('eta>=0')
            return 0
        opt.alphas[j] -= opt.label[j]*(ei-ej)/eta
        opt.alphas[j] = clipa(opt.alphas[j],H,L)
        updateek(opt,j)
        if abs(opt.alphas[j] - jold) < 0.00001:
            print('j not moving enough')
            return 0
        opt.alphas[i] += opt.label[j]*opt.label[i]*(jold - opt.alphas[j])
        updateek(opt,i)
        b1 = opt.b - ei - opt.label[i]*(opt.alphas[i]-iold)*opt.K[i,i]-opt.label[j]*(opt.alphas[j]-jold)*opt.K[i,j]
        b2 = opt.b - ej - opt.label[j]*(opt.alphas[j]-iold)*opt.K[i,j]-opt.label[j]*(opt.alphas[j]-jold)*opt.K[j,j]
        if (0<opt.alphas[i]) and (opt.alphas[i]<opt.C): opt.b = b1
        elif (0<opt.alphas[j]) and (opt.alphas[j]<opt.C): opt.b = b2
        else: opt.b = (b1+b2)/2
        return 1
    else:
        return 0
```

```python
def calcEk(opt,k):
    fxk = np.dot(np.multiply(opt.alphas,opt.label).T,opt.K[:,k])+opt.b #multiply((100,1),(100,1)).T =dot((1,100),100,1) = (1,1)
    Ek = fxk - opt.label[k]
    return Ek
```

```python
def smorbf(datama,labelma,C,toler,num,ktup=('lin',0)):
    opt = Opt(datama,labelma,C,toler,ktup)
    cnt = 0
    entire = True
    achanged = 0
    while(cnt<num) and ((achanged>0) or (entire)):#achanged=0 and entire为False 循环结束
        achanged = 0
        if entire:
            for i in range(opt.m):
                achanged += innerL(i,opt)
                print('fullset,iter:%d i:%d,pairs changed %d' %(cnt,i,achanged))
            cnt +=1
        else:
            nonbou = np.nonzero((opt.alphas>0)*(opt.alphas<opt.C))[0]
            for i in nonbou:
                achanged += innerL(i,opt)
                print('nonbound,iter:%d i:%d pairs changed %d' %(cnt,i,achanged))
            cnt +=1
        if entire: entire = False
        elif (achanged==0): entire=True
        print('iteration number:%d' %cnt)
    return opt.b,opt.alphas
```

```python
def testrbf(datam,labelm,testdata,testlabel,alphas,b,k1):
    svind = np.nonzero(alphas>0)[0]
    #print(svind)
    svs = datam[svind]
    #print(svs)
    labelsv = labelm[svind]
    print('there are %d support vector' %svs.shape[0])
    m,n = datam.shape
    cnt = 0
    for i in range(m):
        kereval = kernelT(svs,datam[i,:],('rbf',k1))
        predict = np.dot(kereval.T,np.multiply(labelsv,alphas[svind]))+b
        if np.sign(predict) != np.sign(labelm[i]):
            cnt +=1
    print('the train error rate: %f' %(float(cnt)/m))
    tcnt = 0
    m,n = testdata.shape
    for i in range(m):
        kereval = kernelT(svs,testdata[i,:],('rbf',k1))
        predict = np.dot(kereval.T,np.multiply(labelsv,alphas[svind]))+b
        if np.sign(predict) != np.sign(testlabel[i]):
            tcnt +=1
    print('the test error rate: %f' %(float(tcnt)/m))    
```

```python
# 这里无法用split()分开，所以采用一个for循环
def img2vector(filename):
    vect = np.zeros((1,1024))
    with open(filename) as fr:
        for i in range(32):
            linestr = fr.readline()
            for j in range(32):
                vect[0,i*32+j] = int(linestr[j])
    return vect
```

```python
import os
def loadimages(dirname):
    hw = []
    filelist = os.listdir(dirname)
    m = len(filelist)
    trainmat = np.zeros((m,1024))
    for i in range(m):
        filearr = filelist[i]
        filestr = filearr.split('.')[0]
        classnum = int(filestr.split('_')[0])
        if classnum==9: hw.append(-1)
        else: hw.append(1)
        trainmat[i,:] = img2vector('%s/%s' %(dirname,filearr))
    return trainmat,hw
    
```
